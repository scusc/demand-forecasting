```python
df_sales.drop(columns=['Secondary Sales Amount STC','Secondary Sales Amount Invoiced','Secondary Sales Amount Returned','Secondary Sales Quantity Returned'],inplace=True)
df_sales
```

```python
df2=pd.read_csv("Test.csv")
df_sales.append(df2,ignore_index=True, sort=False)
df_sales
df2
```

```python
df_sales1 = df_sales.append(df2)
```

```python
df_sales1
```

```python
df_sales1['DATE'] = pd.to_datetime(df_sales1['DATE'])
df_sales1['DATE'] = df_sales1['DATE'].dt.year.astype('str') + '-' + df_sales1['DATE'].dt.month.astype('str') + '-01'
df_sales1['DATE'] = pd.to_datetime(df_sales1['DATE'])
```

```python
df_sales1.rename(columns={"Secondary Sales Quantity Invoiced":"Secondary_Sales_Quantity_Invoiced",'SKU Code':'SKU_Code','CFA Code':'CFA_Code','SKU Combination':'SKU_Combination'},inplace=True)
```

```python
df_sales1 = df_sales1.groupby(['DATE','SKU_Code','CFA_Code','Channel','Quantity','SKU_Combination']).Secondary_Sales_Quantity_Invoiced.sum().reset_index()
```

```python
df_sales1
```

```python
df_diff = df_sales1.copy()
df_diff['Prev_Secondary_Sales_Quantity_Invoiced'] = df_diff['Secondary_Sales_Quantity_Invoiced'].shift(1)
```

```python
df_diff
```

```python
df_diff = df_diff.dropna()
```

```python
df_diff['diff'] = (df_diff['Secondary_Sales_Quantity_Invoiced'] - df_diff['Prev_Secondary_Sales_Quantity_Invoiced'])
```

```python
df_diff
```

```python
df_supervised = df_diff.drop(['Prev_Secondary_Sales_Quantity_Invoiced'],axis=1)
```

```python
for inc in range(1,13):
    field_name = 'lag_' + str(inc)
    df_supervised[field_name] = df_supervised['diff'].shift(inc)
```

```python

df_supervised
```

```python
df_supervised = df_supervised.dropna().reset_index(drop=True)
```

```python
import statsmodels.formula.api as smf 

# Define the regression formula
model = smf.ols(formula='diff ~ lag_1', data=df_supervised)

# Fit the regression
model_fit = model.fit()

# Extract the adjusted r-squared
regression_adj_rsq = model_fit.rsquared_adj
print(regression_adj_rsq)
```

```python
from sklearn.preprocessing import MinMaxScaler
df_model = df_supervised.drop(['DATE','SKU_Code','CFA_Code','Channel','Quantity','SKU_Combination','Secondary_Sales_Quantity_Invoiced'],axis=1)
```

```python
train_set, test_set = df_model[0:-2610].values, df_model[-2610:].values
df_model.info()
```

```python
len(test_set)
```

```python
scaler = MinMaxScaler(feature_range=(-1, 1))
scaler = scaler.fit(train_set)
# reshape training set
train_set = train_set.reshape(train_set.shape[0], train_set.shape[1])
train_set_scaled = scaler.transform(train_set)

# reshape test set
test_set = test_set.reshape(test_set.shape[0], test_set.shape[1])
test_set_scaled = scaler.transform(test_set)
```

```python
X_train, y_train = train_set_scaled[:, 1:], train_set_scaled[:, 0:1]
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
```

```python
X_test, y_test = test_set_scaled[:, 1:], test_set_scaled[:, 0:1]
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])
```

```python
model = Sequential()
model.add(LSTM(4, batch_input_shape=(1, X_train.shape[1], X_train.shape[2]), stateful=True))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X_train, y_train, epochs=1, batch_size=1, verbose=1, shuffle=False)
```

```python
y_pred = model.predict(X_test,batch_size=1)
y_pred
```

```python
y_test
```

```python
y_pred = y_pred.reshape(y_pred.shape[0], 1, y_pred.shape[1])
```

```python

#rebuild test set for inverse transform
pred_test_set = []
for index in range(0,len(y_pred)):
    print (np.concatenate([y_pred[index],X_test[index]],axis=1))
    pred_test_set.append(np.concatenate([y_pred[index],X_test[index]],axis=1))
```

```python
pred_test_set[0]
```

```python
pred_test_set = np.array(pred_test_set)
pred_test_set = pred_test_set.reshape(pred_test_set.shape[0], pred_test_set.shape[2])
```

```python
pred_test_set_inverted = scaler.inverse_transform(pred_test_set)
```

```python
#df_model = df_supervised.drop(['DATE','SKU Code','CFA Code','Channel','Quantity','SKU Combination','Secondary_Sales_Quantity_Invoiced'],axis=1)
result_list = []
sales_dates = list(df_sales1[-2611:].DATE)
sku = list(df_sales1[-2611:].SKU_Code)
cfa = list(df_sales1[-2611:].CFA_Code)
channel = list(df_sales1[-2611:].Channel)
quantity = list(df_sales1[-2611:].Quantity)
skucomb = list(df_sales1[-2611:].SKU_Combination)
act_sales = list(df_sales1[-2611:].Secondary_Sales_Quantity_Invoiced)
for index in range(0,len(pred_test_set_inverted)):
    result_dict = {}
    result_dict['Secondary_Sales_Quantity_Invoiced'] = int(pred_test_set_inverted[index][0] + act_sales[index])
    result_dict['DATE'] = sales_dates[index+1]
    result_dict['SKU Code'] = sku[index+1]
    result_dict['CFA Code'] = cfa[index+1]
    result_dict['Channel'] = channel[index+1]
    result_dict['Quantity'] = quantity[index+1]
    result_dict['SKU Combination'] = skucomb[index+1]
    result_list.append(result_dict)
df_result = pd.DataFrame(result_list)
```
